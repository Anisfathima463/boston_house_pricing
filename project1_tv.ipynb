```python
x=df.iloc[:,:-1]
x
```

```python
y=df.iloc[:,-1]
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=34)
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.preprocessing import StandardScaler
scalar=StandardScaler()
```

```python
x_train=scalar.fit_transform(x_train)
x_test=scalar.fit_transform(x_test)
import pickle
pickle.dump(scalar,open("D:\\project1 tv\\scaling.pkl","wb"))
```

```python
#model training
from sklearn.linear_model import LinearRegression
lm=LinearRegression()
```

```python
lm.fit(x_train,y_train)
```

```python
print(lm.coef_)# multipilers which is multipy to predicted value
```

```python
print(lm.intercept_) #  value of y when x=0,starting point of regression line or plane
```

```python
print(lm.get_params()) #retrieves the hyperparameter of linear regression model
```

```python
#prediction

y_pred=lm.predict(x_test)
y_pred
```

```python
plt.scatter(y_test,y_pred)
```

```python
residuals=y_test-y_pred
```

```python
residuals
```

```python
sns.displot(residuals,kind="kde")
```

```python
plt.scatter(y_pred,residuals)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
```

```python
mse=mean_squared_error(y_pred,y_test)
mse
```

```python
mae=mean_absolute_error(y_pred,y_test)
mae
```

```python
import numpy as np
rmse=np.sqrt(mean_squared_error(y_pred,y_test))
rmse
```
